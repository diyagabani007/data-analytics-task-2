```python
# Import libraries
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Show first 5 rows
print("First 5 rows:")
print(df.head())

# Check dataset info
print("\nDataset Info:")
print(df.info())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# 🔧 Data Cleaning

# 1. Remove duplicates
df.drop_duplicates(inplace=True)

# 2. Rename columns (easy names)
df.rename(columns={
    'math score': 'math_score',
    'reading score': 'reading_score',
    'writing score': 'writing_score'
}, inplace=True)

# 3. Check unique values
print("\nUnique values:")
print(df.nunique())

# 4. Handle outliers (basic example)
df = df[df['math_score'] <= 100]
df = df[df['reading_score'] <= 100]
df = df[df['writing_score'] <= 100]

# 5. Normalize scores (0 to 1 scale)
df['math_score'] = (df['math_score'] - df['math_score'].min()) / (df['math_score'].max() - df['math_score'].min())
df['reading_score'] = (df['reading_score'] - df['reading_score'].min()) / (df['reading_score'].max() - df['reading_score'].min())
df['writing_score'] = (df['writing_score'] - df['writing_score'].min()) / (df['writing_score'].max() - df['writing_score'].min())

# Final cleaned data preview
print("\nCleaned Data:")
print(df.head())

# Save cleaned dataset
df.to_csv("cleaned_students_data.csv", index=False)

print("\n✅ Cleaned dataset saved successfully!")
```
